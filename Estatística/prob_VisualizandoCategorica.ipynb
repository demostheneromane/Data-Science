2.38 Uma pesquisa realizada junto a 1264 mulheres indagou quem seriam seus conselheiros de compras mais confiáveis. Os resultados da pesquisa se deram do seguinte modo : 


|Conselheiros de compras|Percentagem(%)| 
|--|--|
|Propaganda|7|
|Amigos/Familia|45|
|Portal eletronico do fabricante|5|
|Noticiários nos meios de comunicação|11|
|Avaliações de usuários na Internet|13|
|Portais eletronicos de fornecedores|4|
|Vendedores|1|
|Outros|14|

 * Construa um gráfico de barras , um gráfico de pizza e um diagrama de Pareto.

```python


df = { 
    
    "conselheiros_compras" : ["Propaganda","Amigos/familia",
                              "Portal eletronico",
                              "Noticiários de comunicaçoes",
                              "Avaliações de usuários", 
                              "Portais eletronicos de fornecedores",
                              "Vendedores","Outros"], 
    
    "Percentagem" : [7,45,5,11,13,4,1,14], 
    "Percentagem_pareto" : [45,14,13,11,7,5,4,1], # deve se ordenadar
    "conselheiros_compras_pareto" : ["Amigos/Familia","Outros","Avaliações Usuários","Noticiários Comunicação","Propaganda",
                                     "Portal eletronico","Portais Fornecedores","Vendedores"], # mesma ordem de Percentagem_pareo
    "Percentagem_acumulada" : [45,59,72,83,90,95,99,100] # Percentagem acumulada a partir de Percentagem pareto
    
}

#gráfico 1 // gráfico de barras
plt.rcParams["figure.figsize"] = [20,10]
plt.barh(df["conselheiros_compras"],df["Percentagem"])
plt.xlabel("Percentagem (%)")
plt.xlim(0,50)
plt.ylabel("Conselheiros de Compras")
plt.title("Pesquisa de Quem seriam seus conselheiros de compras mais confiáveis",y=1.05)

plt.show()

```

```python

#gráfico 2 // gráfico pizza

plt.pie(df["Percentagem"],labels=["Propaganda 7%","Amigos/Familia 45%","Portal eletronico do fabricante 5%", 
                                  "Noticiários nos meios de comunicação 11 %","Avaliações de usuários na Internet 13%",
                                  "Portais  eletronicos de fornecedores 4%","Vendedores 1%","Outros 14%"])

plt.title("Pesquisa de Quem seriam seus conselheiros de compras mais confiáveis",y=1.0)
plt.show()
```

```python
#gráfico de Pareto

fig,ax = plt.subplots(figsize=(26,13))
ax.set_ylim(0,100)
ax1 = ax.bar(df["conselheiros_compras_pareto"],df["Percentagem_pareto"])
ax2 = ax.twinx()
ax2.plot(df["conselheiros_compras_pareto"],df["Percentagem_acumulada"],marker="s")
ax2.set_ylim(0,100)
plt.show()
```

* Que método gráfico vc imagina que seja o melhor para retratar esses dados ? 
  
  
  rep: O gráfico de barras. 
  



2.39 O que vc faria se ganhasse um 1milhão ? Em uma pesquisa de opinião da March Madness NCAA, foi realizado, um levantamento junto a 1.078 adultos , a quem se perguntou em que gastariam inicialmente o dinheiro caso viessem a ganhar 1 milhão. Os resultados se apresentaram do seguinte modo . 


|Opção de Gasto|Percentagem(%)|
|--|--|
|Comprar ingressos para a 2010 Final Four|4%|
|Caridade|4%|
|Pagar dividas|59%|
|Poupar|16%|
|Fazer um cruzeiro|8%|
|Pegar o dinheiro, jamais se inscrever em uma gincana novamente|4%|
|Tentar a sorte em Las Vegas|5%|

* Construa um gráfico de barras , um gráfico de pizza e um diagrama de Pareto . 

```python
df = {
    "opção de gasto" : ["Pagar dividas","Poupar","Fazer um cruzeiro","Tentar/Las vegas" ,
                        "Caridade","Comprar ingressos","Pegar dinheiro"], 
    
    "Percentagem" : [59,16,8,5,4,4,4],
    "Percentagem_acumulada" : [59,75,83,88,92,96,100]
}

#gráfico de barras 


plt.barh(df["opção de gasto"],df["Percentagem"])
plt.title("O que vc Faria com um 1Milhão ? ",y=1.05)
plt.xlabel("Percentagem(%)")
plt.ylabel("Opção de gasto")
plt.show()
```

```python
#gráfico de pizza
plt.pie(df["Percentagem"],labels=["Pagar dividas","Poupar","Fazer um cruzeiro","Tentar/Las Vegas","Caridade",
                                  "Comprar ingressos","Pegar dinheiro"])

plt.title("O que vc Faria com um 1 Milhão ? ", y=1.05)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.set_ylim(0,100)
ax.set_title("O que vc Faria com um 1Milão ?",y=1.05)
ax.set_xlabel("Opcão de Gasto",labelpad=20)
ax.set_ylabel("Percentagem %")
ax1 = ax.bar(df["opção de gasto"],df["Percentagem"])
ax2 = ax.twinx()
ax2.plot(df["opção de gasto"],df["Percentagem_acumulada"],marker="s")
ax2.set_ylim(0,100)
plt.show()
```

* Que método gráfico vc imagina que seja o melhor para retratar esses dados ? 

  rep: O gráfico de Pareto
