**O Diagrama de Pareto** 
Em um diagrama de Pareto , os resultados de cada uma das categorias são inseridos em um gráfico sob a forma de barras verticais , em ordem descendente , de acordo com suas respectivas frequências , e combinados com uma linha de percentagens acumuladas o mesmo gráfico .O Diagrama de Pareto consegue revelar situações nas quais ocorre o principio de Pareto.


   O PRINCÌPIO DE PARETO 
   O principio de Pareto existe quando a maioria dos itens em um determinado conjunto de dados ocorre em uma pequena quantidade de categorias e os poucos itens remanescentes estão dispersos ao longo de uma grande quantidade de categorias . Esses dois grupos são geralmente chamados de "poucos vitais" e "muitos triviais".
   
   
Um diagrama de Pareto apresenta a capacidade de separar os "poucos vitais" dos "muitos triviais" , possibilitando que voce se concentre nas categorias importantes.

**O Histograma** 


Um histograma é um gráfico de barras para dados numéricos agrupados no qual voce utiliza barras verticais para representar as frequencias ou percentagens em cada grupo . Voce exibe a variável de interesse ao longo do eixo horizontal (X) . O eixo vertical (Y) representa a frequencia ou a percentagem de valores por intervalo de classe.

**O Polígono de Percentagens** 


Caso tentasse construir dois ou mais histogramas no mesmo gráfico , voce não conseguiria interpretar facilmente cada um dos histogramas , já que as barras iriam se sobrepor . Quando existem dois ou mais grupos , voce deve utilizar um polígono de percentagens . Um polígono de percentagens utiliza os pontos médios de cada um dos intervalos de classe para representar os dados de cada das classes e , depois disso aponta no gráfico os pontos médios em suas respectivas percentagens de classe sob a forma de pontos em uma linha.

**FIGURA 2.14** 


Polígonos de percentagens para o custo de refeições em restaurantes no centro da cidade e em restaurantes no subúrbio.


```python
df= {
    "custo_refeicao" : [20,30,40,50,60,70,80], 
    "percentagem_cidade" : [2,18,30,18,18,12,2], 
    "percentagem_suburbio" : [0,24,26,26,18,6,0]
}
fig ,ax = plt.subplots(figsize=[10,6])
ax.set_xlim(10,80)
ax.set_ylim(0,35)
ax.set_xlabel("Custo da Refeição ($)")
ax.set_ylabel("Percentagem")
ax.set_title("Polígonos de Percentagens para o custo de refeições em Restaurantes no centro da cidade e em Restaurantes no Subúrbio",y=1.05)
ax1 = ax.plot(df["custo_refeicao"],df["percentagem_cidade"],marker="x",label="Cidade")
ax2 = ax.plot(df["custo_refeicao"],df["percentagem_suburbio"],marker="o",label="Subúrbio")
plt.legend()
plt.show()
```

Uma análise dos dois polígonos da Figura 2.14 leva á conclusão de que a maior concentração do custo de refeições em restaurantes do centro da cidade se encontra entre 30 e 40 (dollars) , enquanto o custo de refeições em restaurantes do subúrbio se encontra de maneira uniforme entre 20 e 50(dollars).

**O Polígono de Percentagens Acumuladas(Ogiva)** 

O poligono de percentagens acumuladas , ou ogiva , utiliza a distribuição de percentagens acumuladas discutida na Seção 2.4 para exibir a variável de interesse ao longo do eixo X e as percentagens acumuladas ao longo do eixo Y.

**O Gráfico de Séries Temporais** 


Um gráfico de séries temporais insere os valores de uma varíavel no eixo Y, e insere no eixo X o periodo de tempo associado a cada valor numérico. Um gráfico de séries temporais pode ajudar a explorar tendencias nos dados que ocorrem ao longo do tempo.Por exemplo , considere a quantidade de energia saolar instalada(em megawatts) nos Estados  Unidos , por ano , ao longo do periodo de 2000 até 2008.
