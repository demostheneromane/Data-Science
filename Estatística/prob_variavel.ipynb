2.1 Tres (3) diferentes tipos de bebidas são vendidos em uma lanchonete - refrigerantes , chá e café. 

   * Explique por que o tipo de bebida vendida é um exemplo de uma variável categórica 
     
     rep : O tipo de bebida vendida é um exemplo de uma variável categórica porque ele tem um conjunto de valores que podem ser 
     posicionados em categorias tais como : refrigerantes, chá e café 
     ex: 
     |produto|Tipo de bebida vendida| 
     |--|--|
     |Coca cola|refrigerantes|
     |Camomila|chá|
     |Starbucks|café|
     
   * EXplique por que o tipo de bebida vendida é um exemplo de uma variável de escala nominal. 
   
     rep : O tipo de bebida vendida é um exemplo de uma variável de escala nominal porque os diferentes categorias tais como : 
     refrigerantes, chá e café são classificados em categorias distintas e não leva em conta as diferenças entre as categorias. 
     
     
   
   
2.2 Em uma lanchonete são vendidos refrigerantes de tres (3) tamanho - pequeno , médio e grande . Explique por que o tamanho do refrigerante é um exemplo de uma variável de escala ordinal . 

rep : O tamanho do refrigerante é um exemplo de uma variável de escala ordinal porque os diferentes categorias do tamanho tais como : pequeno , médio e grande são classificados em categorias distintas e leva em conta as diferenças entre as categorias (pequeno, médio e grande) .
