* Construa um histograma e um poligono de percentagens. 


```python
tarifas_energia_eletrica = [82,90,95,96,102,108,109,111,114,116,
                            119,123,127,128,129,130,130,135,137,
                            139,141,143,144,147,148,149,149,150,
                            151,153,154,157,158,163,165,166,167,
                            168,171,172,175,178,183,185,187,191,
                            197,202,206,213]

plt.rcParams["figure.figsize"] = [20,10]
plt.hist(tarifas_energia_eletrica,bins=[20,40,60,80,100,120,140,160,180,200,220])
plt.xlabel("Tarifas de Energias Eletrica")
plt.ylabel("Frequencia")
plt.title("Custo de energia elétrica durante o mes de julho 2009",y=1.05)
plt.ylim(0,15)
plt.show()
```

```python
df={
    "tarifas_energia_eletrica" : [40,60,80,100,120,140,160,180,200,220], # intervalo de classe de Dados Brutos sobre Tarifas de 
     "percentagem_energia"       : [0,0,0,8,14,18,26,18,10,6]              # serviços de Energia Eletrica 
    
}
fig,ax =plt.subplots()
ax.plot(df["tarifas_energia_eletrica"],df["percentagem_energia"],marker='s')
ax.set_xlabel("Tarifas de Energias Eletrica")
ax.set_ylabel("Percentagem")
ax.set_title("Custo de energia elétrica durante o mes de julho 2009",y=1.05)
ax.grid(axis='y')
for y,x in enumerate(df['tarifas_energia_eletrica']): 
    ax.annotate(x,xy=(x+2.5,df['percentagem_energia'][y]))
plt.show()
```

* Construa um poligono de percentagens acumuladas 

```python
df= {
     "tarifas_energia_eletrica" : [40,60,80,100,120,140,160,180,200,220],
     "percentagem_acumulada_energia" : [0,0,0,8,22,40,66,84,94,100]
}

plt.rcParams["figure.figsize"] = [13,6]
plt.plot(df["tarifas_energia_eletrica"],df["percentagem_acumulada_energia"],marker='s')
plt.xlabel("Tarifas de Energias Eletrica")
plt.ylabel("Percentagem (%)")
plt.title("Custo de energia elétrica durante o mes de julho 2009",y=1.05)

for pY,pX in enumerate(df['tarifas_energia_eletrica']):
    plt.annotate(pX,xy=(pX+2.5,df["percentagem_acumulada_energia"][pY]))
   
plt.grid()
plt.show()

```
